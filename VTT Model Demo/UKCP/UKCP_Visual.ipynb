### Cleaned temperature and humidity UKCP data projection for South East England 2020-2024

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
df = pd.read_csv('UKCP_2040_VTT.csv')
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
```

First 5 Years Plot

```python
# Filter the data for the first five years
df_5years = df[df.index.year <= 2024]

# Plot
fig, ax = plt.subplots(figsize=(12,6))
color = 'tab:red'
ax.plot(df_5years.index, df_5years['Mean_Temperature'], color=color)
ax.set_xlabel('Date')
ax.set_ylabel('Temperature', color=color)
ax.tick_params(axis='y', labelcolor=color)

ax2=ax.twinx()
color = 'tab:blue'
ax2.set_ylabel('Humidity', color=color)
ax2.plot(df_5years.index, df_5years['Mean_Humidity'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

# Show the plot
plt.show()
```

All 20 Years Plot
